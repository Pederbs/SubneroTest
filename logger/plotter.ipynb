```python
import pandas as pd
import matplotlib.pyplot as plt

file = 'SUMS_logger_current2.csv'
df = pd.read_csv(file)

df['Current'] = df['Current']/100
df['Power'] = df['Current']*df['Voltage']

df.plot(x='Time', y=['Voltage', 'Current'])
```

# Logging av Strøm med Spenning som referanse

RAW_V_A er vanlig konfigurasjon
RAW_A_V er A0 og A1 byttet på hverandre 

ser ikke ut som om det gjorde noe forskjell
ser også ut som at strømmåleren ikke fungerer


```python
file_VA = 'SUMS_logger_RAW_V_A.csv'
df_VA = pd.read_csv(file_VA)

file_AV = 'SUMS_logger_RAW_A_V.csv'
df_AV = pd.read_csv(file_AV)

df_big = df_VA
df_big['Voltage_A1'] = df_AV['Voltage']*1
df_big['Current_A0'] = df_AV['Current']*1

df_big.plot(x='Time', y=['Voltage', 'Voltage_A1'])
```

```python
df_big.plot(x='Time', y=['Current', 'Current_A0'])
```
